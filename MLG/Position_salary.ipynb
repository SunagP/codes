```python
import numpy as np
import pandas as pd
```

```python
data  = pd.read_csv('Position_Salaries.csv')
X = data.iloc[:,1:2].values
y = data.iloc[:,-1].values
print(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, 
                                                     random_state = 0)
```

```python

from sklearn.linear_model import LinearRegression
regressoragent = LinearRegression()
regressoragent.fit (X_train, y_train )            
```

```python
import matplotlib.pyplot as plt
predictValues = regressoragent.predict(X_test)

print(y_test - predictValues)

plt.scatter(X_train, y_train, color = 'green')
plt.plot (X_train, regressoragent.predict (X_train), color = 'red')
plt.title ('compare Training result - Area/Price')
plt.xlabel('Area of Flat')
plt.ylabel('Price')
plt.show()


plt.scatter(X_test, y_test, color = 'green')
plt.plot (X_train, regressoragent.predict (X_train), color = 'red')
plt.title ('compare Test result - Area/Price')
plt.xlabel('Area of Flat')
plt.ylabel('Price')
plt.show()


plt.scatter(X_train, y_train, color = 'green')
plt.scatter(X_test, y_test, color = 'blue')
plt.plot (X_train, regressoragent.predict (X_train), color = 'red')
plt.title ('compare ENTIRE result - Area/Price')
plt.xlabel('Area of Flat')
plt.ylabel('Price')
plt.show()
```

```python

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4,include_bias = False)
X_train_trans = poly_reg.fit_transform(X_train)
X_test_trans = poly_reg.fit_transform(X_test)
print(X_test_trans)
linregressor2 = LinearRegression()
linregressor2.fit(X_train_trans, y_train)
```

```python
val = [[68]]
predictionlinear = regressoragent.predict (val)
predictionPoly = linregressor2.predict (poly_reg.fit_transform(val))

print ("linear prediction = ", predictionlinear)
print ("polynomial prediction = ", predictionPoly)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, linregressor2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
